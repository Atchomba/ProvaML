```python
import pandas as pd
import numpy as np
```

```python
url = "https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/immobili.csv"

df = pd.read_csv(url)
df
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop('zn', axis=1, inplace=True)
```

```python
df.drop('chas', axis=1, inplace=True)
```

```python

df.rename(columns={'medv': 'price'}, inplace=True)
```

```python
df
```

```python
df.drop(columns=['Total_Zeros',	'Total_Zeri'], inplace=True)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

   
```

```python
X = df.drop(columns='price')
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2, 
                                                    random_state=667)
```

```python

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_

	
```

```python
y_pred =model.predict(X_test)
```

```python
r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print('R2_score: ', r2score)
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
```

```python
import joblib
joblib.dump(model,'Nuovomodel.pkl')
```

```python
model_loaded = joblib.load('Nuovomodel.pkl')
model_loaded
```

```python
print('coeficente:' , model_loaded.coef_)
```

```python
print('interception:', model_loaded.intercept_)
```
